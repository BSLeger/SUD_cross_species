Disease Page:

Field|	Descriptions
| --- | --- |
Term Name|	The Disease Ontology (DO)term and its unique accession ID, linked to its page at the Disease Ontology web site.
Definition|	The full DO description for the selected term
Synonyms|	Alternative search terms for the selected term
Cross References|	Alternative IDs from additional disease vocabularies, linked when possible to the relevant source|
Parent Terms|	The immediate parent terms for the selected disease, linked to their Alliance disease pages. Because the Disease Ontology is built on a DAG framework, a structure may have more than one parent term.
Child Terms|	Disease terms below the page’s term in the ontology hierarchy
Sources of Associations|	Links to the appropriate Disease Ontology pages at the following Model Organism Databases (MODs). The presence of a link means the database has a disease page for the disease, but does not indicate whether disease models are present. By default, the table shows 10 rows per page. Use the toggle beneath the table to increase the number of rows.

Associations:
Field|	Descriptions
| --- | --- |
Gene/Allele/Model|	A gene/allele/model associated with the disease, linked to the Alliance gene page for the species
Species|	The species for the gene associated with the disease
Association Type|	How the gene is associated with the disease (e.g. “is implicated in,” meaning that some variant of the gene is shown to function in causing or modifying a disease (for human) or a disease model state. “A marker of” refers to a biological molecule found in blood, other body fluids, or tissues that is a sign of a the disease.)
Disease|	The Disease Ontology subterms. Subterms (Subtypes) are linked to their Alliance Disease page.
Evidence|	The supporting data for the disease association. Mouse over an Evidence Code for a description of the code.
Based On|	The orthologous gene whose data was used for the disease association
Source|	The Model Organism Database (MOD) that curated the gene to disease association, linked to the source database
References|	The PubMed ID for the publication that was used to make the gene to disease association, linked to PubMed. If the association was made through a database load, the MOD database ID is shown, linked to the MOD database record.
