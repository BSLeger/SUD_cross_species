```r
library(GenomicSEM)
library(here)
```

```r
getwd()
rename_files=FALSE
```

```r
setwd('/tscc/projects/ps-palmer/brittany/SUD_cross_species/')
munged.dir <- here("/tscc/projects/ps-palmer/brittany/SUD_cross_species/nicsa_munge/nicsa_munge_wchr")
ld.dir <- here("/tscc/projects/ps-palmer/brittany/SUD_cross_species/nicsa_ldsc/ld_scores_wchr")
wld.dir <- here("/tscc/projects/ps-palmer/brittany/SUD_cross_species/ldsc/ld_scores_wchr")
```

```r
#rename files to make them function in LDSC package
rename files=FALSE
ldscore_files <- list.files(ld.dir, full.names = TRUE)
#remove ldsc_5Mb_chr from file name
new_names <- substring(basename(ldscore_files), 13)
new_file_paths <- file.path(dirname(ldscore_files), new_names)
if (rename_files)
    {file.rename(ldscore_files, new_file_paths)}
```

```r
#define files to be analyzed
trait_tbl=read.table(here('nicsa_munge_ref/nicsa_traits.csv'),sep=',',header=1)
trait_names=sort(trait_tbl$file_prefix)
traits <- sort(list.files(munged.dir, pattern = "\\.gz$", full.names=TRUE))

sample.prev <- rep(NA, length(traits))
population.prev <- rep(NA, length(traits))
```

```r
trait_names=trait_names[trait_names!= 'regressedlr_nicsa_day1_infusion']
traits=traits[traits!='/tscc/projects/ps-palmer/brittany/SUD_cross_species/nicsa_munge/nicsa_munge_wchr/regressedlr_nicsa_day1_infusion.sumstats.gz']
```

```r
LDSCoutput <- ldsc(traits = traits, sample.prev = sample.prev, population.prev = population.prev, ld = ld.dir, wld = wld.dir, trait.names = trait_names, chr = 20)

save(LDSCoutput, file="nicsa_LD_score_regression/LDSCoutput_rerun.RData")
```

```r
write.table(LDSCoutput$S, "nicsa_LD_score_regression/S.tsv",sep='\t')
write.table(LDSCoutput$V, "nicsa_LD_score_regression/V.tsv",sep='\t')
write.table(LDSCoutput$I, "nicsa_LD_score_regression/I.tsv",sep='\t')
write.table(LDSCoutput$N, "nicsa_LD_score_regression/N.tsv",sep='\t')
write.table(LDSCoutput$m, "nicsa_LD_score_regression/m.tsv",sep='\t')
```

```r
sample.prev <- rep(NA, length(traits))
population.prev <- rep(NA, length(traits))

LDSCoutput <- ldsc(traits = traits, sample.prev = sample.prev, population.prev = population.prev, ld = ld.dir, wld = wld.dir, trait.names = trait_names, chr = 20)

save(LDSCoutput, file="nicsa_LD_score_regression/LDSCoutput_rerun.RData")
```
