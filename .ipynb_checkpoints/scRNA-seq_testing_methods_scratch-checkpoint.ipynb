```python
# Function to calculate effect size (Cohen's d)
def cohen_d(x1, x2):
    n1, n2 = len(x1), len(x2)
    print(f'{n1}, {n2}')
    #n needs to change for 
    s1, s2 = np.var(x1, ddof=1), np.var(x2, ddof=1)
    pooled_sd = np.sqrt((s1 / n1) + (s2 / n2))
    return (np.mean(x1) - np.mean(x2)) / pooled_sd
```
